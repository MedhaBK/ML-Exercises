#Pandas Exercises

```python
import pandas as pd
```

```python
data = pd.read_csv('Salaries.csv')
```

```python
data.head()
```

```python
#Using the info method to find out how many entries there are:
data.info()
```

```python
#Taking the average of BasePay col:

# data['BasePay'].mean() : this will give error as not all values are numbers here so first we need to convert them to numbers:

data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
data["BasePay"].describe()
```

```python
print(data["BasePay"].mean())
```

```python
#Maximum amount of OvertimePay:
data['OvertimePay'] = pd.to_numeric(data['OvertimePay'], errors='coerce')
data['OvertimePay'].max()
```

```python
# What is the job title of JOSEPH DRISCOLL ?
data[data["EmployeeName"] == "JOSEPH DRISCOLL"]["JobTitle"]
```

```python
#How much does JOSEPH DRISCOLL make (including benefits)?

data[data['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
# Name of highest paid person:

name = data["TotalPayBenefits"].max()
data['EmployeeName'][data["TotalPayBenefits"] == name]
```

```python
#Lowest paid person :
data['TotalPayBenefits'].min()

#The strange thing is his TotalPayBenefits amount is negative
```

```python
#What was the average (mean) BasePay of all employees per year? (2011-2014) ?

data.groupby('Year')[data.select_dtypes(include='number').columns].mean()['BasePay']
```

```python
#Unique job titles:

data['JobTitle'].nunique()
```

```python
#Top 5 common jobs:

data['JobTitle'].value_counts()
data['JobTitle'].value_counts().head()
```

```python
#How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)

import numpy as np
values = data[data['Year'] == 2013]["JobTitle"].value_counts() == 1
print(np.sum(values == True))
```

```python
#People with word "Chief" in their job title:

sum(data['JobTitle'].str.contains("Chief"))
```

```python
#Bonus question (Correlation):

# data["JobTitle"].corr(data["TotalPayBenefits"])
data["JobTitle"].str.len().corr(data["TotalPayBenefits"])
```
