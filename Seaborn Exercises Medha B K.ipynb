```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

titanic = titanic.dropna(subset = ['fare','age'])

sns.jointplot(x='fare',y='age',data = titanic)
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

sns.histplot(x = 'fare', data = titanic, color = 'red', bins = 18)
plt.xlim(0,titanic['fare'].max())
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

titanic['fare'].max()
```

```python
sns.boxplot(x = 'class', y = 'age', data = titanic, palette = ['lightblue','lightgreen','orange'])
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

sns.swarmplot(x='class',y = 'age', data = titanic, palette = ['green','orange','blue'])
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

sns.countplot(x = 'sex', data=titanic, palette = ['blue','green'])
plt.grid(True, axis='y', linestyle='-', alpha=0.5)
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

plt.figure(figsize=(8,6))
sns.heatmap(titanic.corr(numeric_only = True), cmap = 'coolwarm', center = 0)
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!


plot = sns.FacetGrid(titanic, col="sex")
plot.map(plt.hist, "age", bins=10, color="lightblue", edgecolor="black")
plt.grid(True, axis='y', linestyle='-', alpha=0.5)

plt.show()
```
